```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

dataNf = pd.read_csv('result_benchN.csv')
dataZf = pd.read_csv('result_benchZ.csv')
dataMf = pd.read_csv('result_bench_monad.csv')
dataSOf = pd.read_csv('result_bench_so.csv')
series = [dataNf, dataZf, dataMf, dataSOf]
data = pd.concat(series, axis=1, join='inner')

_ = plt.boxplot(series, labels=('FuSe', 'FuSe_unsafe', 'FuSe_monad', 'SO'))
data.describe()
```
